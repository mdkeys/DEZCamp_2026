```python
from pyspark.sql import SparkSession
#SparkSession tells how we connect to spark.
```

```python
spark = SparkSession.builder \
    .master("local[*]") \
    .appName('test') \
    .getOrCreate()

# Connect to local master (thing for coordinating jobs of spark cluster. 
# This will create a local cluster and [*] means use all available CPUs.
# You can change * to a number to limit the number of CPUs

#appName will be 'test'. 
#getOrCreate will get it if the spark session exists, or create it if it does not.
```

```python
# We see that it doesn't recognize the first row is column names
# Add option to change that:
df = spark.read.option("header","true") \
    .csv('taxi_zone_lookup.csv')

df.show()
```
