```python
# Use sqlalchemy to connect to postgresql 
# Need to create pgsql container first before this step
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
# Get data schema:
print(pd.io.sql.get_schema(df_green, name='green_tripdata', con=engine))
print(pd.io.sql.get_schema(df_zones, name='taxi_zones', con=engine))
```

```python
#Create header for the data
df_green.head(n=0).to_sql(name='green_tripdata', con=engine, if_exists='replace') 
#df_green.head(0) # shows just the header of the data
```

```python
# Add tqdm to monitor chunk load process
!uv add tqdm
```

```python
import tqdm
print(tqdm.__version__)
```

```python
from tqdm.auto import tqdm
```

```python
# Create iteration script to load df_green (in chunks) & df_zones to pgcli (pgsql cli)
# Note: The example in the training does not work because of options limitations with read_parquet.

# Ingest df_green in chunks to pgsql
df_green.head(0).to_sql(name='green_tripdata', con=engine, if_exists="replace")
total_rows = len(df_green)
chunksize = 100000

for i in tqdm(range(0, total_rows, chunksize)):
    chunk = df_green.iloc[i:i + chunksize]
    chunk.to_sql(name='green_tripdata', con=engine, if_exists="append")

print('Done with green_tripdata')
```

```python
# Now ingest df_zones in chunks to pgcli
df_zones.to_sql(name='taxi_zones', con=engine, if_exists="replace")
print('Done with zones data')
```
